```python
class Neural_Network(object):
    def __init__(self, Lambda=0):        
    
        self.inputLayerSize = 3
        self.outputLayerSize = 1
        self.hiddenLayerSize = 6
        
       
        self.W1 = np.random.randn(self.inputLayerSize,self.hiddenLayerSize)
        self.W2 = np.random.randn(self.hiddenLayerSize,self.outputLayerSize)
        
        #Regularization Parameter:
        self.Lambda = Lambda
        
    def forward(self, X):
        #Propogate inputs though network
        self.z2 = np.dot(X, self.W1)
        self.a2 = self.stable_softmax(self.z2)
        self.z3 = np.dot(self.a2, self.W2)
        yHat = self.stable_softmax(self.z3) 
        return yHat
        
#Done Softmax        
    def stable_softmax(self, z):
        exps = np.exp(z - np.max(z))
        return exps / np.sum(exps)

    

    def sigmoid(self, z):
        #Apply sigmoid activation function to scalar, vector, or matrix
        return 1/(1+np.exp(-z))

    
    def costFunction(self, X, y):
        #Compute cost for given X,y, use weights already stored in class.
        self.yHat = self.forward(X)
        J = 0.5*sum((y-self.yHat)**2)/X.shape[0] + (self.Lambda/2)*(np.sum(self.W1**2)+np.sum(self.W2**2))
        return J
    
    def cross_entropy(self, X, y):

        m = y.shape[0]
        p = stable_softmax(self.X)
  
        log_likelihood = -np.log(p[range(m),y])
        loss = np.sum(log_likelihood) / m
        return loss
        
    def delta_cross_entropy(self, X, y):

        m = y.shape[0]
        grad = stable_softmax(self.X)
        grad[range(m),y] -= 1
        grad = grad/m
        return grad
    
    
    def costFunctionPrime(self, X, y):

        self.yHat = self.forward(X)
        
        delta3 = np.multiply(-(y-self.yHat), self.softmaxPrime(self.z3))

        dJdW2 = np.dot(self.a2.T, delta3)/X.shape[0] + self.Lambda*self.W2
        
        delta2 = np.dot(delta3, self.W2.T)*self.softmaxPrime(self.z2)

        dJdW1 = np.dot(X.T, delta2)/X.shape[0] + self.Lambda*self.W1
        
        return dJdW1, dJdW2
    

    def getParams(self):

        params = np.concatenate((self.W1.ravel(), self.W2.ravel()))
        return params
    
    def setParams(self, params):
  
        W1_start = 0
        W1_end = self.hiddenLayerSize*self.inputLayerSize
        self.W1 = np.reshape(params[W1_start:W1_end], \
                             (self.inputLayerSize, self.hiddenLayerSize))
        W2_end = W1_end + self.hiddenLayerSize*self.outputLayerSize
        self.W2 = np.reshape(params[W1_end:W2_end], \
                             (self.hiddenLayerSize, self.outputLayerSize))
        
    def computeGradients(self, X, y):
        dJdW1, dJdW2 = self.delta_cross_entropy(X, y)
        return np.concatenate((dJdW1.ravel(), dJdW2.ravel()))
```
