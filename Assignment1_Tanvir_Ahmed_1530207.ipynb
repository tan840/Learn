```python
class Neural_Nets():
    def __init__(self, Lambda=0): 
        
        
        self.inputLayers = 3
        self.outputLayer = 1
        self.hiddenLayer = 6
        
        #weights
        self.W1 = 0.10 * np.random.randn(self.inputLayers,self.hiddenLayer)
        self.W2 = 0.10 * np.random.randn(self.outputLayer,self.hiddenLayer)
        self.Lambda = Lambda
        
    def Feedforward(self, X):
        #Calculate the dot product between inputs and weights
        self.z1 = np.dot(X, self.W1)
        
        #Pass the result from step 1 through an activation function)
        self.l1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.l1, self.W2.T)
        yHat = self.sigmoid(self.z2)
        #print(yHat)
        return yHat
    
    def sigmoid(self, z):
        return 1/(1+np.exp(-z))
    
    
    def sigmoid_der(self,z):
        return self.sigmoid(z) *(1-self.sigmoid (z))
    
    def LossFunction(self,x,y):
        self.y_pred = self.Feedforward(x)
        squared_errors = (y - self.y_pred) ** 2
        return squared_errors
    
    def BackPropagate(self,y_hat,y):
        self.squared_errors = self.LossFunction(x,y)
        
        self.output_layer = self.Feedforward(x)
        
        l2_delta = self.squared_errors* self.sigmoid_der(self.output_layer)
        
        l1_error = l2_delta.dot(self.W2)
        
        l1_delta = l1_error*self.sigmoid_der(np.dot(x, self.W1))
        
        #update the weights
        
        self.W2 +=  np.dot(l2_delta.T,self.l1)
        self.W1 += x.T.dot(l1_delta)
        
    
        #print(self.W1)
        #print(self.W2)
        
    def TrainNN(self):
        for interation in range(10):
            self.BackPropagate(self.Feedforward(x),y)
            
        
       
    def testData(self):
        #Test Data

        T_x = pd.DataFrame(np.random.rand(1000, 3), columns = ['x1', 'x2', 'x3'])

        T_y = np.array([np.random.randint(2, size = 1000)])
        T_y = T_y.reshape(1000, 1)
        T_x = T_x.to_numpy()
        
        print(T_y)
        print(T_x)
        for interation in range(100):
            self.BackPropagate(self.Feedforward(T_x),T_y)
        
        p =self.Feedforward(T_x)
        print(p)
        
        plt.plot(T_y)
        plt.plot(p)
        plt.grid(1)
        plt.xlabel('X')
        plt.ylabel('y')
    
```
